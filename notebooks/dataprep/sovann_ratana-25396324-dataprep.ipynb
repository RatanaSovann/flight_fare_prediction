+ Rationale: same routes should have the same totalDistance

```python
# Fill missing values of total distance by the median totalDistance of th3 same route
# Fill missing 'totalTravelDistance' by grouping over 'startingAirport', 'destinationAirport', and 'segmentsDepartureAirportCode'
df_copy['totalTravelDistance'] = df_copy.groupby(['startingAirport', 'destinationAirport', 'segmentsDepartureAirportCode'])['totalTravelDistance'].transform(
    lambda x: x.fillna(x.median())
)
```
